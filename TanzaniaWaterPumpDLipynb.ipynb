This is part 1 of a 4 part series in which I test the preformance of deep learning models against more simple approaches. In this dataset we are working on the Tanzania water pumps problem in which we are able to predict which water pumps will need repair so we can properly deploy workers.

```
from potosnail import DeepLearning, DataHelper
import pandas as pd
import numpy as np

dl = DeepLearning()
dh = DataHelper()

df = pd.read_csv('https://raw.githubusercontent.com/spe301/ML-vs-DL/main/data/Tanzania_Water_Pumps/train_df_final.csv')
```

# Data Preprocessing

scaling data and converting to an array. I'm also one hot encoding the labeles, so that the model can process it.

```
X = np.array(dh.ScaleData('standard', df.drop(['label'], axis='columns')))
y = dl.MulticlassOutput(np.array(df['label']))
```

```
X.shape, y.shape
```

# Modeling

This function runs a single fold gridsearch and records the model's preformance data along with model hyperparameters.

This code took _ to run with a GPU

```
params = {'output_dim': [3], #we have 3 possible outputs
          'nodes': [128, 256], 
          'activation': ['relu', 'tanh'], 
          'regularizer': [None, 'L1', 'L2'], 
          'stacking': [False, True], #making the first hidden layer a duplicate of the input layer
          'dropout': [False, True], 
          'nlayers': [3], #number of layers excluding stacking and closer (so realy 3-5 layers)
          'closer': [False, True], # adding an extra layer before output
          'loss_func': ['categorical_crossentropy', 'sparse_categorical_crossentropy'], #also testing out sparse because we hale a lot of zeroes
          'optimizer': ['adam'] #also known as cost function
          }

adam = dl.CollectPerformance(params, dl.DeepTabularClassification, X, y, epochs=150, batch_size=64)
adam.loc[adam['val_accuracy']==max(adam['val_accuracy'])]
```

```
adam = pd.read_csv('https://raw.githubusercontent.com/spe301/ML-vs-DL/main/data/Tanzania_Water_Pumps/Results/adam.csv')
adam.loc[adam['val_accuracy']==max(adam['val_accuracy'])]
```

```
params = {'output_dim': [3], #we have 3 possible outputs
          'nodes': [128, 256], 
          'activation': ['relu', 'tanh'], 
          'regularizer': [None, 'L1', 'L2'], 
          'stacking': [False, True], #making the first hidden layer a duplicate of the input layer
          'dropout': [False, True], 
          'nlayers': [3], #number of layers excluding stacking and closer (so realy 3-5 layers)
          'closer': [False, True], # adding an extra layer before output
          'loss_func': ['categorical_crossentropy', 'sparse_categorical_crossentropy'], #also testing out sparse because we hale a lot of zeroes
          'optimizer': ['rmsprop'] #also known as cost function
          }

rms = dl.CollectPerformance(params, dl.DeepTabularClassification, X, y, epochs=150, batch_size=64)
rms.loc[rms['val_accuracy']==max(rms['val_accuracy'])]
```

```
#rms.to_csv('drive/MyDrive/rms.csv')
```

```
rms = pd.read_csv('https://raw.githubusercontent.com/spe301/ML-vs-DL/main/data/Tanzania_Water_Pumps/Results/rms.csv')
rms.loc[rms['val_accuracy']==max(rms['val_accuracy'])]
```

```
results = pd.concat([adam, rms])
```
